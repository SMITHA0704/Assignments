```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the cars dataset to python environment
data=pd.read_csv(r'C:\Users\Smitha.s1_oob\OneDrive - GEMS Education\Documents\PERSONAL\DS\cars_data.csv')
print(data)
```

```python
#Check for the null values present in the dataset
print(data.isnull().values.any())
```

```python
#Displaying the columns
print(data.columns)
```

```python
#Plot a bar graph of male vs female buyers participated in the sales.
import numpy as np
x=np.arange(1)
f=data[data['Buyer Gender']=='Female']['Buyer Gender'].count()
print(f)
m=data[data['Buyer Gender']=='Male']['Buyer Gender'].count()
print(m)
plt.bar(x,m,width=0.50,color='r',label='Male buyers count')
plt.bar(x+0.50,f,width=0.50,color='k',label='Female buyers count')
plt.legend()
plt.xticks(x+0.25,['Gender'])
plt.title('Count of Male vs. Female buyers')
plt.show()
```
