```python
#Importing data
import pandas as pd
df=pd.read_csv('/content/Sales_data_zone_wise.csv')
print(df)
```

```python
#The mean sales generated by each zone.
print(df.mean())
```

```python
#Total sales generated by all the zones for each month.
data=df.groupby(['Month'])['Zone - A','Zone - B','Zone - C','Zone - D'].sum()
print(data)
```

H0-All zones generate same amount of sales
H1-All zones do not generate same amount of sales

```python
#To Check whether all the zones generate the same amount of sales.
import scipy.stats as stats
fvalue,pvalue=stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
print(fvalue,pvalue)
if pvalue<0.05:
    print('Ample evidence to reject the null hypothesis')
else:
    print('Ample evidence to fail to reject the null hypothesis')
```
