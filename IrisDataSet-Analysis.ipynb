```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the dataset to python environment
data=pd.read_csv(r'C:\Users\Smitha.s1_oob\OneDrive - GEMS Education\Documents\PERSONAL\DS\iris.csv')
print(data)
```

```python
#Display the columns in the dataset.
print('The columns of iris dataset=')
print(data.columns)
```

```python
#Calculate the mean of each column of the dataset
print(data.mean())
```

```python
#Check for the null values present in the dataset
print(data.isnull().values.any())
```

```python
#Visualization-1
#Comparison between iris classifications based on sepal length and width
import seaborn as sns
plt.title('Comparison based on sepal length and width')
sns.scatterplot(data['SL'],data['SW'],hue =data['Classification'],s=50)
```

```python
#Visualization-2
#Comparison between various species based on petal length and width
plt.title('Comparison between various species based on petal length and width')
sns.scatterplot(data['PL'],data['PW'], hue = data['Classification'], s= 50)
```

```python
#Visualization-3
#boxplot to show how the iris classification is distributed with all four input variables
fig, axes = plt.subplots(2, 2,figsize=(16,9))
sns.boxplot( y='PW', x='Classification', data=data, orient='v' , ax=axes[0, 0])
sns.boxplot( y='PL', x='Classification', data=data, orient='v' , ax=axes[0, 1])
sns.boxplot( y='SL', x='Classification', data=data, orient='v' , ax=axes[1, 0])
sns.boxplot( y='SW',x='Classification', data=data, orient='v' , ax=axes[1, 1])
plt.show()
```
